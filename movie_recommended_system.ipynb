```python
import pandas as pd 
import numpy as np
```

# Data Gathering

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies.shape
```

```python
## Remove unecessary Feature 
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
```

```python
movies.info()
```

```python
movies=movies[['movie_id','overview','title','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
```

```python
def converted (text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(converted)
```

```python
movies['keywords']=movies['keywords'].apply(converted)
```

```python
movies.head()
```

```python
def con(text):
    l=[]
    c=0
    for i in ast.literal_eval(text):
        if c<3:
            l.append(i['name'])
        c+=1
    return l
```

```python
movies['cast']=movies['cast'].apply(con)
```

```python
movies.head()
```

```python
def director(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
    return l
```

```python
movies['crew']=movies['crew'].apply(director)
movies.head()
```

```python
def space(text):
    l=[]
    for i in text:
        l.append(i.replace(' ',''))
    return l
```

```python
movies['cast']=movies['cast'].apply(space)
movies['genres']=movies['genres'].apply(space)
movies['crew']=movies['crew'].apply(space)
movies['keywords']=movies['keywords'].apply(space)
movies.head()

```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['cast']
```

```python
movies.head()
```

```python
new=movies.drop(['genres','overview','keywords','cast','crew'],axis=1)
```

```python
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x:' '.join(x))
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```
